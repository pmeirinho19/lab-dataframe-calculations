## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here

#SUBPROB1

clients_total_spent = orders.groupby(['CustomerID']).agg({'amount_spent':sum})
clients_total_spent
```

```python
orders['CustomerID'].nunique() #--> so the aggregation went well
```

```python
#SUBPROB2

q95 = round(clients_total_spent['amount_spent'].quantile(q=0.95),2)
q95 #--> the clients with a spent amount greater than q95 will be labeled as VIP
q75 = round(clients_total_spent['amount_spent'].quantile(q=0.75),2)
q75 # --> clients with total spents between q75 and q95 will be labeled as Preferred
```

```python
#SUBPROB3

vip_clients_df = clients_total_spent[clients_total_spent['amount_spent'] > q95]
vip_clients_df
```

```python
vip_clients_df.nunique() # --> to check if the group by went well. it's ok since the name of rows = nunique
```

```python
preferred_clients_df = clients_total_spent[(clients_total_spent['amount_spent'] < q95) & (clients_total_spent['amount_spent'] > q75)]
preferred_clients_df
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here

list_vip = list(vip_clients_df.index) # to store the vip id's

vip_orders_df = orders[orders['CustomerID'].isin(list_vip)] # to see all the orders of vip clients
vip_orders_df
```

```python
# now is time to see the country with most vip clients
country_vip = vip_orders_df.groupby(['Country']).agg({'CustomerID': 'nunique'})
country_vip # So UK is the country with the greater number of VIP clients.
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here

# first we need to to know the country with greater number of preferred clients
    # for this the method is the same as the one used in for vip clients
    
list_preferred = list(preferred_clients_df.index) # to store the preferred id's

preferred_clients_orders_df = orders[orders['CustomerID'].isin(list_preferred)] # to see all the orders of vip clients

country_preferred = preferred_clients_orders_df.groupby(['Country']).agg({'CustomerID': 'nunique'})
country_preferred
```

```python
a = country_vip
b = country_preferred

countrys_vip_preferred = pd.merge(left = a, right = b, left_on = 'Country', 
                                                   right_on = 'Country', how = 'inner' )
countrys_vip_preferred

# we could do another column with the sum from the count of CustomerID, but we can see clearly
    # that the country with the greater number of combined customers(vip+preferred) is also UK
```
